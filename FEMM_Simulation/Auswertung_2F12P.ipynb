```python
import matplotlib
# matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
import pandas as pd
import math as m
import numpy as np
```

```python
file0='/home/kalap/Kai/MasterETTUWien/Studium_Aktiv/Masterarbeit_Wind/FEM/APSM/kont_2F12P/2F12P_0DegA/2F12P_V2_0Deg_results.txt'
#file0="/home/kalap/Kai/MasterETTUWien/Studium_Aktiv/Masterarbeit_Wind/FEM/APSM/kont_2F12P/test000/2F12P_V2_0Deg_results.txt"
df0=pd.read_csv(file0,sep=";",skipinitialspace=True)
file0K007='/home/kalap/Kai/MasterETTUWien/Studium_Aktiv/Masterarbeit_Wind/FEM/APSM/kont_2F12P/2F12P_0K007DegA/2F12P_V2_0K007Deg_results.txt'
#file0K002="/home/kalap/Kai/MasterETTUWien/Studium_Aktiv/Masterarbeit_Wind/FEM/APSM/kont_2F12P/test007/2F12P_V2_0Deg_results.txt"
df007=pd.read_csv(file0K007,sep=";",skipinitialspace=True)
```

```python
frames=[df0,df007]
labels=['0°','0,007°']
#frames=[df0];labels=['0'];
li_names=['mu', 'mo', 'r1u', 'r1o', 'r2u','r2o', 'r3u', 'r3o', 'l1u', 'l1o', 'l2u', 'l2o', 'l3u', 'l3o']
li_names_u=['mu','r1u', 'r2u', 'r3u', 'l1u','l2u','l3u']
li_names_o=['mo','r1o', 'r2o', 'r3o', 'l1o','l2o','l3o']
```

```python
df0.columns
for df in frames:
    display(df[li_names].head(2))
```

```python
fig, ax1 =plt.subplots(figsize=(15,8))
for [df,label] in zip(frames,labels):
    for li in li_names_u:
        ax1.plot(df['i'],df[li]/1E-7,label=(label+' '+li),lw=1)
#plt.xlim(90,270)
#plt.ylim(-2.3e-6,-1.2e-6)
plt.grid()
ax1.legend(bbox_to_anchor=(1,1), loc="upper left")
ax1.set_ylabel(r'$ \Phi$ 10^(-7) [Wb]')
ax1.set_xlabel('Position in Grad')
plt.title("Fluss durch ein Längenelement der Statorspule")
#plt.savefig("Fluss1.png", bbox_inches='tight',dpi=900)
plt.show()
```

```python
for df in frames:
    df['phi_u']=df['mu']+7/8*(df['l1u']+df['r1u'])+3/8*(df['l1u']+df['r1u'])+1/8*(df['l1u']+df['r1u'])
    df['phi_o']=df['mo']+7/8*(df['l1o']+df['r1o'])+3/8*(df['l1o']+df['r1o'])+1/8*(df['l1o']+df['r1o'])
    df['phi']=1/2*(df['phi_u']+df['phi_o'])
    df['phi_diff']=df['phi'].diff()
    df['phi_m']=1/2*(df['mu']+df['mo'])
    df['phi_m_diff']=df['phi_m'].diff()
    df['phi_mu_diff']=df['mu'].diff()
    df['phi_mo_diff']=df['mo'].diff()
```

```python
fig, ax1 =plt.subplots(figsize=(15,8))
for [df,label] in zip(frames,labels):
    for li in li_names_u:
        if(label==labels[1]):
            ls='--'
        else:
            ls='-'
        ax1.plot(df['i'],df[li]/1E-7,label=(label+' '+li),lw=1,ls=ls)
#plt.xlim(90,270)
plt.ylim(-25,25)
plt.grid()
ax1.legend(bbox_to_anchor=(1,1), loc="upper left")
ax1.set_ylabel(r'$ \Phi$ 10^(-7) [Wb]')
ax1.set_xlabel('Position in Grad')
plt.title("Fluss durch ein Längenelement der Statorspule")
#plt.savefig("Fluss1.png", bbox_inches='tight',dpi=900)
plt.show()
```

```python
fig, ax1 =plt.subplots(figsize=(15,5))
for [df,label] in zip(frames,labels):
    ax1.plot(df['i' ],df['mu']/1e-7,label=label,lw=1)
plt.xlim(90,130)
#ax1.set_ylim(20,120)
#plt.grid()
ax1.legend(bbox_to_anchor=(1,1), loc="upper left")
ax1.set_ylabel(r'$ \Phi$ E-7 [Wb]')
ax1.set_xlabel('Position in Grad')
plt.title("Fluss durch ein Längenelement der Statorspule")
#plt.savefig("Fluss1.png", bbox_inches='tight',dpi=900)
plt.show()
```

```python
fig, ax1 =plt.subplots(figsize=(15,8))
for [df,label] in zip(frames,labels):
    for li in ['phi_mu_diff','phi_mo_diff']:
        if(label==labels[1]):
            ls='--'
        else:
            ls='-'
        ax1.plot(df['i'],df[li]/1E-7,label=(label+' '+li),lw=1,ls=ls)
#plt.xlim(90,270)
#plt.ylim(-25,25)
plt.grid()
ax1.legend(bbox_to_anchor=(1,1), loc="upper left")
ax1.set_ylabel(r'$ \Phi$ 10^(-7) [Wb]')
ax1.set_xlabel('Position in Grad')
plt.title("Fluss durch ein Längenelement der Statorspule")
#plt.savefig("Fluss1.png", bbox_inches='tight',dpi=900)
plt.show()
```
