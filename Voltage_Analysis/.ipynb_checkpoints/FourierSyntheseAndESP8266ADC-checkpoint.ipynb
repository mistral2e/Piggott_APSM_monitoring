## Fourier Synthese
* by using the binned spektrum with a bin with of 1Hz


```python
# peak Voltage faktors from ./RotorDeviation_URMS_nconst.ipynb 
n1=3;up1=27.914491*2**(1/2)
n2=6;up2=28.274522*2**(1/2)
```

```python
kc=range(0,100)
f0=31.0;N=5000
T0=1/f0/p
x = np.linspace(0.0, T0, N, endpoint=False)
fig, axo = plt.subplots(1,figsize=(15,5))
ax=axo
for [n,up] in [[n1,up1],[n2,up2]]:
    y_ges=np.zeros(N)
    for k in kc:
        amp=spek_bins_xk.loc[spek_bins_xk['xk']==(k),n].iloc[0]
        y = np.sin(k*f0 *2*np.pi*x)*amp
        y_ges+=y
        #display(k,amp)
    y_ges=y_ges/y_ges.max()*up
    ax.plot(y_ges,label='Aufnahme '+str(n)+' '+scope_dat.iloc[n]['Symbol'])
    if(n==n1):
        y_ges_n1=y_ges.copy()
    if(n==n2):
        y_ges_n2=y_ges.copy()
ax.axhline(y=0,color='grey',ls='--')
ax.legend(loc='lower left')
ax2=ax.twinx()
ax2.plot((y_ges_n2-y_ges_n1),c='g',label='Difference: ('+str(n2)+' '+scope_dat.iloc[n2]['Symbol']+' - '+str(n1)+' '+ scope_dat.iloc[n1]['Symbol']+')')
ax2.legend(loc='upper right')
ax.set_title("Fourier Synthese")
ax.set_ylabel("Voltage [V]")
ax2.set_ylabel("Difference Voltage [V]")
ax2.set_ylim(-0.8,0.8)
plt.show()
```
