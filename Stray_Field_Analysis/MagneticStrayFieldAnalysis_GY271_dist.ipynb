# 3D Distance
Measure Distance betweeen correct (gerade) and faulty (schräg) rotor position

```python
#Example
nodes = random.randint(10, size=(2, 3))
node=(0, 0, 0)

dist = distance.cdist([node], nodes, 'euclidean')
closest_index=dist.argmin()

display(nodes)
display(dist)
display(closest_index)
display(dist[0,closest_index])
#Probe:
#c^2=x^2+y^2+z^2
display((nodes[closest_index,0]**2+nodes[closest_index,1]**2+nodes[closest_index,2]**2)**(1/2))
```

```python
def mean_distance_nearest_points(df_ref,df_test):
    sumA=0.0
    for index, row in df_test.iterrows():
        node=[row['Bx'],row['By'],row['Bz']]
        dist = distance.cdist([node], df_ref[['Bx','By','Bz']], 'euclidean')
        closest_index=dist.argmin()
        sumA=sumA+dist[0,closest_index]
    return sumA/index
```

## Distance between nearest points of correct position
* Conclusion: The mean distance between measured gyroscope datapoints arises when the postion of the sensor (and vice versa rotor) is altered.

```python
mean_distance_nearest_points(df_gerade,df_gerade_2rounds1)
```

```python
mean_distance_nearest_points(df_gerade,df_gerade_2rounds2)
```

```python
mean_distance_nearest_points(df_gerade_2rounds1,df_gerade_2rounds2)
```
